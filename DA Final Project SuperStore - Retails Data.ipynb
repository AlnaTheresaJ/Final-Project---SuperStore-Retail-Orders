#Handling Missing Values

```python
df.isnull().sum()
#The 'df.isnull().sum' function is used to identify the total number of missing(null) values in each column of thr dataset.
#Handling missing data is a critical step in data preprocessing.Ignoring null values can lead to misleading insights or errors during analysis.
#So,by analysing the output,it is evident that every column has zero value.This means that,no missing values are present in the entire dataset.
```

```python
#Drop rows where critical fields are missing
df.dropna(subset=['Order Date','Product ID','Quantity'], inplace = True)
#This command removes rows where any of the critical fields like - 'Order Date','Product ID','Quantity'-containg missing(NaN) values.
#'subset' specifies the columns to check for nulls.
#'inplace=True' tells pandas to apply the changes directly to the original DataFrame without creating a nw one.
```

#Removing Duplicates

```python
df.drop_duplicates(inplace=True)
#It checks for duplicate rows across all columns and helps to maintain a clean dataset for reliable analysis and modeling.
#This command removes exact duplicate rows from the dataset,ensuring that each entry is unique and valid.
```

#Correcting Datatypes

```python
# 1 Convert to Datetime Format
#To ensure accurate time-based analysis,such as calculating shipping delays, analysing monthly sales trends,it is essential that all dete-related columns are in datetime format.
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%b-%y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%b-%y')
```

```python
# 2 Convert Numeric Columns to float or int
#To ensure that all numerical columns in the dataset are stored in the correct numeric datatypes(float or int) enabling accurate calculations,statistical analysis and visualizations.
df['Sales'] = pd.to_numeric(df['Sales'])
df['Profit'] = pd.to_numeric(df['Profit'])
df['Discount'] = pd.to_numeric(df['Discount'])
df['Shipping Cost'] = pd.to_numeric(df['Shipping Cost'])
df['Unit Price'] = pd.to_numeric(df['Unit Price'])
df['Quantity'] = pd.to_numeric(df['Quantity'])
df['Row ID'] = pd.to_numeric(df['Row ID'])
```

```python
# 3 Convert Categorical Columns to catergory
#This is done to convert string-based columns to category datatype inorder to optimize memory usage and improve performance.
#Also, provides benefits like faster computation and more clarity in analysis.
#Ideal for grouping and pivoting.
cat_cols = [
    'Segment','Ship Mode','Gender','Region','Category','Sub-Category',
    'Order Priority','Customer ID','Customer Name','Product Name'
]

for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')
```

#Creating Derived Columns

```python
#Derived Column 1 - Shipping Delay Days
#Derive a new column named 'Shipping Delay Days' to measure the time delay between when an order was placed and when it was shipped. 
#This column allows to evaluate time-based performance,shipping performance comparison across different ship modes and to identify whether shipping speed impacts sales or profitability.
df['Shipping Delay Days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
#Derived Column 2 - Profit Margin Percent
#Deriva a column named 'Profit Margin Percent' to calculate the profit margin percent for each individual order to understand the profitability relative to sales.
#This column gives a clear idea:
#To compare the profitability across orders or groups(product,category,region).
#To identify high margin products or customers.
#To spot loss-making transactions.
#Profit Margin % = Profit/Sales x 100
df['Profit Margin Percent'] = (df['Profit']/df['Sales']) * 100
```

```python
#Derived Column 3 - Discount %
#Derive a column named 'Discount %' to express the discount values as a percentage for easier understanding,readability and comparison.
#This transformation enhances clarity when analysing how much discount was applied to each order.
#Discount % = Discount x 100
df['Discount %'] = df['Discount'] * 100
```

#Filtering Data

```python
#1 Filter and display orders with a discount greater than 30%
#This filters dataset df to include onlt those orders where the discount given was greater than 30%.
#The result is stored in a new DataFrame called High_discount_Orders.
High_Discount_Orders = df[df['Discount'] > 0.3]
```

```python
#2 Filter orders with shipping delays greater than 3 days
#This filters the dataset df to only include orders that were sipped late specifically,where the shipping delay was moree than 3 days.
#Shipping Delay Days is assumed to be a column that shows how many days late the shippment was.
#The result is stored in a new DataFrame called Late_shipments.This delay could significantly impact customer satisfaction or retention.
Late_Shipments = df[df['Shipping Delay Days'] > 3]
```

```python
#17 Correlation Matrix - Heatmap
#To identify and understand the linear relationship between numerical variables - Sales,Profit,Quantity and Discount using a correlation matrix visualized through a heatmap.
#This helps to assess how change in one variable might be assosciated with the changes in another.
#Sales vs Profit : Higher sales almost always leads to higher profit.Profit is closely driven by sales volume and has a very strong positive correlation.This suggests that growing sales is a reliable way to grow profits.
#Sales vs Quantity : Sales increases with quantity,but not perfectly.This indicates a moderate positive correlation.
#Quantity vs Discount : Discounts dont significantly impact quantity sold.So there is no correlation.
#Giving discounts does not significantly increase the number of items sold.
corr = df[['Sales','Profit','Quantity','Discount']].corr()
```

```python
corr
```

```python
#The heatmap shows the correlation matrix between four variables : Sales,Profit,Quantity and Discount.
#Red indicates strong positive correlation(closer to +1).
#Blue indicates negative correlation(closer to -1).
#White/light color shows weak or no correlation(close to 0).
#Sales & profit shows very strong positive correlation(correlation = +1)
#Sales & Quantity shows moderate positive correlation(correlation = +44)
#Profit &Quantity shows moderate positive correlation(correlation = +43)
#Discount & Profit shows weak negative correlation(correlation = -16)
#Discount & Sales shows weak negative correlation(correlation = -16)
#Discount & Quantity shows almost no correlation(correlation = +0.0081)
#The heatmap highlights that sales and profit are strongly linked,while discounts have a weak or negative on both sale s and profit.This indicates that 
#Profitability is best achieved by focusing on sales growth,optimising product mix and maintaining proper margins,rather than relying on discount.
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.figure(figsize=(10,5))
plt.show();
```
